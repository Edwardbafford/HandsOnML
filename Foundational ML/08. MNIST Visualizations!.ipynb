## MNIST Visualizations
Have some fun plotting the digit dataset with some dimensionality reduction algorithms

```python
import numpy as np
from sklearn.datasets import fetch_mldata

# Get MNIST data

idx = np.random.permutation(60000)[:10000]

mnist = fetch_mldata('MNIST original')
X = mnist['data'].astype(np.float)[idx]
y = mnist['target'][idx]
```

```python
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt 

# Visualize with TSNE

tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(13,10))
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y)
plt.axis('off')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Visualize with PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(13,10))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y)
plt.axis('off')
plt.colorbar()
plt.show()
```

Wow!! TSNE was waaaay better than PCA!!
