## Decision Tree on Moons
Pretty simple.. train a decision tree on the moons dataset from Scikit-learn

```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

#Collect and split data

X, y = make_moons(n_samples=10000, noise=.4)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

#Train model

params = {
    'max_depth':[2,4,8,20],
    'min_samples_split':[2,5,10,20,40],
    'min_samples_leaf':[1,2,5,10,20],
    'max_leaf_nodes':[2,5,10,100,None]
}

clf = GridSearchCV(DecisionTreeClassifier(), params, cv=3)
clf.fit(X_train, y_train)
```

```python
clf.best_estimator_
```

```python
from sklearn.metrics import accuracy_score

#View Performance!

y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```
